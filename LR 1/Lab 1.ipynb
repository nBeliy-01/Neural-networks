<h1>NumPy</h1>

```python
import numpy as np
```

```python
arr = np.zeros((10,10))
for i in range(len(arr)):
    arr[i] = np.random.randint(0,15,10)
print("Вихідна матриця:\n", arr)
print("Розмірність: ", arr.shape)
```

<h3>Операції з числами</h3>

```python
print("Масив + число:\n", arr + 10)
print("\nМасив - число:\n", arr - 10)
print("\nМасив * число:\n", arr * 10)
print("\nМасив / число:\n", arr / 10)
```

<h3>Операції з векторами</h3>

```python
vec = np.arange(10, 30, 2)
print("Вихідний вектор:\n", vec)
print("Розмірність", vec.shape)
```

```python
print("Масив + вектор:\n", np.add(arr, vec))
print("\nМасив - вектор:\n", np.subtract(arr, vec))
print("\nМасив * вектор:\n", np.dot(arr, vec), 
      "\nРозмірність результуючого вектора: ", np.dot(arr, vec).shape)
print("\nМасив / вектор:\n", np.divide(arr, vec))
```

<h3>Операції з матрицями</h3>

```python
matrix = np.zeros((10, 10))
for i in range(len(matrix)):
    matrix[i] = np.random.randint(20, 25, 10)
print("Вихідна матриця:\n", matrix)
print("Розмірність: ", matrix.shape)
```

```python
print("Матриця + матриця:\n", np.add(arr, matrix))
print("\nМатриця - Матриця:\n", np.subtract(arr, matrix))
print("\nМатриця * Матриця:\n", np.dot(arr, matrix), 
      "\nРозмірність результату: ", np.dot(arr, matrix).shape)
print("\nМатриця / Матриця:\n", np.divide(arr, matrix))
```

<h1>Matplotlib</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
```

```python
x = np.arange(-10, 10, 1)
y_true = np.array([y for y in np.arange(10, 50, 2)])
y = np.array([y + np.random.rand()*np.random.randint(-5, 5, 1) for y in np.arange(10, 50, 2)])

fig = plt.figure()
ax = fig.add_subplot(111)

ax.set(title = 'Лінійна функція з похибкою',
       xlim = [5, 60],
       xlabel = 'Вісь X',
       ylabel = 'Вісь Y')

plt.grid(True)
plt.plot(y_true, x, color='blue', label='True values')
plt.scatter(y, x, color='red', label='Generated values')
plt.legend()

box = {'facecolor':'lightblue',
       'edgecolor':'black'}

plt.text(45, -5.0, 'mAE = ' + str(round(mae(y_true, y), 3)), bbox=box)
plt.text(45, -7.5, 'mSE = ' + str(round(mse(y_true, y), 3)), bbox=box)

plt.show()
```

```python
df = pd.DataFrame({
    'X' : x,
    'Y_true' : y_true,
    'Y_gen' : y.transpose()[0],
     'mAE': mae(y_true, y),
    'mSE': mse(y_true, y)
})
df.to_csv('output_linear_function.csv')
```
